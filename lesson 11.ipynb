```python
# create a class:
# use pass to create an empty class in Python without causing a syntax error
class Cat:
    pass

print(Cat) # Cat is defined in the main module
```

```python
# to add information to Cat:
# self is not a keyword, you can change it to other words you like, but self is a naming convention 
class Cat:
    def __init__(self, name, age): # __init__ initializer method
        self.name = name # creates insntance attribute: "name"
        self.age = age

# _init_() is a method, __init__(self, name, age) belongs to class Cat now

```

```python
Oreo = Cat("Oreo", 1)
print(Oreo)
print(Oreo.name)
print(Oreo.age)
Whiskers = Cat("Whiskers", 5)
print(Whiskers.name)
print(Whiskers.age)
```

```python
# _str_: Human-Readable Representation
class CatCls:
    Home = "Tampa"  # Class attribute
    def __init__(self, name, age):
        self.name = name  # Instance attribute
        self.age = age    # Instance attribute
    def __str__(self):
        return f"Cat(name={self.name}, age={self.age}, home={CatCls.Home})"

# create an instance of Cat
Cookie = CatCls("Cookie", 5)
# to demonstrate the __str__ method defined in the CatCls, resulting in a more readable output
print(Cookie)  # This will use the __str__ method
print(Whiskers) # output: the class name and the memory address of the object, the address will change if I re-run the previous block
```

```python
# child and parent class
# parent class
class Animal:
    def __init__(self, name):
        self.name = name  # Instance attribute
    def speak(self):
        return "Some sound"

# child class that inherits from Animal
class Dog(Animal):
    def speak(self):  # Override the parent method
        return "Woof!"

# Define another child class
class Cat(Animal):
    def speak(self):  # Override the parent method
        return "Meow!"

# creat instance of parent class
dog = Animal("Puppy")
# create instances of the child classes
Buddy = Dog("Buddy")
Belly = Cat("Belly")

# Call methods from the instances
print(dog.name)
print(dog.speak())

print(Buddy.name)  
print(Buddy.speak())

print(Belly.name) 
print(Belly.speak())
```

```python
class CatCls:
    Home = "Tampa"  # Class attribute
    def __init__(self, name, age):
        self.name = name  # Instance attribute
        self.age = age    # Instance attribute
    def __str__(self):
        return f"Cat(name={self.name}, age={self.age}, home={CatCls.Home})"
class CatShelter(CatCls):
    def __init__(self, name, age, phone):  # add a new initializer, and overs the parent's (self, name, age)
        self.name = name
        self.age = age  
        self.phone = phone
Cookie = CatCls("Cookie", 5)
Mittens = CatShelter("Mittens", 3, "(727) 946-0000")
print(Mittens.name)
print(Mittens.age)
print(Mittens.phone)
print(Cookie)
print(Mittens) # inherit from CatCls' Home, and __str__ method
```
